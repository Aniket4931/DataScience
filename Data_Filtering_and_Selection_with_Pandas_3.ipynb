# Data Filtering and Selection with Pandas

#### Problem: Use Pandas to load a dataset containing information about movies, including columns: 'Title', 'Genre', 'Year', and 'Rating'. Write a Python script to:


```python
import pandas as pd
```

```python
movies_data=pd.read_csv("Data/movies_data.csv")
```

```python
movies_data
```

#### Filter the DataFrame to select movies released after 2010 with a rating above 8.

```python
filter_movies=movies_data[(movies_data["Year"]> 2010)&(movies_data["Rating"]>8)]
```

```python
filter_movies
```

#### Display the top 10 movies based on their ratings.

```python
sorted_movies=movies_data.sort_values("Rating")
```

```python
Ten_movies=sorted_movies.tail(10)
```

```python
Ten_movies
```
