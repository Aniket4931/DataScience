## Problem: Given a CSV file containing sales data with columns: 'Date', 'Product', 'Units Sold', and 'Revenue', write a Python script using Pandas to:
me.

```python
import pandas as pd
```

### Read the data into a DataFrame.


```python
sales_data=pd.read_csv("Data/sales_data.csv")
```

```python
sales_data.head()
```

## Calculate the total revenue generated each day.


```python
a=0
for Revenue,row in  sales_data.iterrows():
    a+=row["Revenue"]
print("Total",a)
```

## Plot a line graph showing the daily revenue over time.

```python
import matplotlib.pyplot as plt
```

```python
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
daily_revenue = sales_data.groupby('Date')['Revenue'].sum()
plt.plot(daily_revenue.index, daily_revenue.values)
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```
